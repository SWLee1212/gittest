```python
from gurobipy import *

time_for_work = [[14,5,8,7],[2,12,6,5],[7,8,3,9],[2,4,6,10]]
machine_for_job_Var = [ ["x"+str(a)+str(b) for b in range(4)] for a in range(4)]

m = Model("Assignment problem")
x_variables = []
for a in range(4):
    x_variables.append([])
    for b in range(4):
         x_variables[a].append(
             m.addVar(
                vtype = GRB.BINARY,
                obj = time_for_work[a][b], 
                name="(%s)" % (machine_for_job_Var[a][b])
             ))
            
    
m.modelSense = GRB.MINIMIZE
m.update()

m.getObjective()            

for machine in range(4):
    machine_constraions = [x_variables[machine][job] for job in range(4)]    
    m.addConstr(
        quicksum(machine_constraions) == 1,
        "Machine %s Constraints" % machine )

for job in range(4):
    job_constraions = [x_variables[machine][job] for machine in range(4)]    
    m.addConstr(
        quicksum(machine_constraions) == 1,
        "job %s Constraints" % job)
    
m.update()
m.getConstrs()      

m.optimize()

for v in m.getVars():
    print(v.varName,v.x)
    
print(m.getObjective().getValue())    
```
