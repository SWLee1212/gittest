```python
from gurobipy import *

m = Model("Assignment Problem")

NUMBER_OF_MACHINES = 4
NUMBER_OF_JOB = 4

cost_matrix = [
    [14,5,8,7],[2,12,6,5],[7,8,3,9],[2,4,6,10]
]

variables = [
    [m.addVar(vtype = GRB.BINARY,
              obj = cost_matrix[machine][job],
              name = "(Machine-%d, Job-%d)" % (machine+1, job+1))            
    for job in range(NUMBER_OF_JOB)]
        for machine in range(NUMBER_OF_MACHINES)
]

m.modelSense = GRB.MINIMIZE
m.update()

m.getObjective()  

for machine in range(NUMBER_OF_MACHINES):
    m.addConstr(
        quicksum( variables[machine][job] for job in range(NUMBER_OF_JOB)) == 1,
        "Job requirement for machine %d" % (machine+1)
    )
for job in range(NUMBER_OF_JOB):
    m.addConstr(
        quicksum( variables[machine][job] for machine in range(NUMBER_OF_MACHINES)) == 1,
        "Job requirement for machine %d" % (job+1)
    )    
    
m.optimize()

for v in m.getVars():
    print(v.varName,v.x)
    
print(m.getObjective().getValue())        
```
